```python
# 4. Null value analysis
print("\nMissing values:\n", df.isnull().sum())
```

```python
#  Drop or fill nulls
# Step 1: Check outliers using IQR
Q1 = df['duration'].quantile(0.25)
Q3 = df['duration'].quantile(0.75)
IQR = Q3 - Q1
lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR
outliers = df[(df['duration'] < lower) | (df['duration'] > upper)]

# Step 2: Decide
if len(outliers) > 0.05 * len(df):  # if >5% are outliers
    fill_value = df['duration'].median()
    print("Outliers detected â†’ Using median")
else:
    fill_value = df['duration'].mean()
    print("No significant outliers â†’ Using mean")

# Step 3: Fill
df['duration'].fillna(fill_value, inplace=True)
df
```

```python
fill_value = df['language'].mode()[0]
df['language'].fillna(fill_value, inplace=True)
```

```python
fill_value = df['collaboration'].mode()[0]
df['collaboration'].fillna(fill_value, inplace=True)
```

```python
# Top genres or artists
if 'genre' in df.columns:
    print("\nTop genres:\n", df['genre'].value_counts())

if 'artist' in df.columns:
    print("\nTop artists:\n", df['artist'].value_counts())
```

```python
# Popularity distribution
if 'popularity' in df.columns:
    plt.figure(figsize=(10, 5))
    sns.histplot(df['popularity'], kde=True)
    plt.title("Track Popularity Distribution")
    plt.xlabel("Popularity Score")
    plt.ylabel("Count")
    plt.show()
```

```python
top_artists = df['artist'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_artists.values, y=top_artists.index, palette='viridis')
plt.title('Top 10 Artists by Track Count')
plt.xlabel('Number of Tracks')
plt.ylabel('Artist')
plt.show()
```

```python
top_genres = df['genre'].value_counts().head(5)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_genres.values, y=top_genres.index, palette='rainbow')
plt.title('Top 5 by Genres')
plt.xlabel('Number of Tracks')
plt.ylabel('Genres')
plt.show()
```

```python
languages = df['language'].value_counts().head(3)

plt.figure(figsize=(12, 6))
sns.barplot(x=languages.values, y=languages.index, palette='viridis')
plt.title('Top 3 languages')
plt.xlabel('Number of Tracks')
plt.ylabel('languages')
plt.show()
```

```python
top_songs = df['song_title'].value_counts().head(10)

plt.figure(figsize=(12, 6))
sns.barplot(x=top_songs.values, y=top_songs.index, palette='rainbow')
plt.title('Top 10 biggest hit songs')
plt.xlabel('Number of Tracks')
plt.ylabel('top_songs')
plt.show()
```
